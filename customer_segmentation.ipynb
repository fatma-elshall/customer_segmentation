**RFM analysis**

```python
import datetime as dt
```

```python
#the last date of buying
df['InvoiceDate'].max()
```

```python
#select date after to analysis on
today_date=dt.datetime(2011,12,11)
```

**make RFM function **
which contain three basic steps :

1- date of selling
2-frequancy : number of selling times
3-sum for total price

```python
#rfm function
rfm=df.groupby('CustomerID').agg({'InvoiceDate' :lambda date:(today_date-date.max()).days,
                                  'InvoiceNo' :lambda num : num.nunique(),
                                 'TotalPrice' : lambda TotalPrice: TotalPrice.sum() })
rfm
```

```python
#change culumn names
rfm.columns=['Recency' ,'Frequancy' ,'Monetary']
rfm
```

recency , frequancy ,monitary these are values which i will build analysis of data on them .

```python
#select values that monetary greater than zero
rfm=rfm[rfm['Monetary'] >0]
rfm
```

```python
rfm.describe().T
```

```python
#make score for receny , frequancy and monetary to make numbers easy understand
rfm['recency_score']=pd.qcut(rfm['Recency'],5 ,labels=[5,4,3,2,1])
```

```python
rfm['frequancy_score']=pd.qcut(rfm['Frequancy'].rank(method='first'),5 ,labels=[1,2,3,4,5])
```

```python
rfm['monetary_score']=pd.qcut(rfm['Monetary'],5 ,labels=[1,2,3,4,5])
```

```python
rfm.sample(10)
```

```python
#make rfm score code for customer
#make concatination for recency score and frequancy score to use in rfm_score code to categorize customers

rfm['RFM_score']=(rfm['recency_score'].astype(str)+rfm['frequancy_score'].astype(str))
rfm.head()
```

```python
#segmentation customers using RFM score
seg_map={
    r'[1-2][1-2]' : 'hibernating',
    r'[1-2][3-4]' : 'at_risk',
    r'[1-2]5' : 'cant_loose',
    r'3[1-2]' : 'about_to_sleep',
    r'33' : 'need_to_attention',
    r'[3-4][4-5]' : 'loyal_customers',
    r'41' : 'promissing ',
    r'51' : 'new_customers',
    r'[4-5][2-3]' : 'potential_loyalists',
    r'5[4-5]' : 'champions'

}

rfm['segment']=rfm['RFM_score'].replace(seg_map , regex=True)
rfm.sample(10)
```

this is the main table we want .

```python
rfm[['segment' ,'Recency','Frequancy','Monetary']].groupby('segment').agg(['mean','count','max']).round()
```
